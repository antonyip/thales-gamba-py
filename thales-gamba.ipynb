```python
from flipside import Flipside
import requests
import json
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
```

```python
resp = requests.get("https://api.flipsidecrypto.com/api/v2/queries/88f07287-590c-441e-aca5-0b17ba42714a/data/latest")
df = pd.DataFrame(resp.json())
print(df.tail())

settled = requests.get("https://api.flipsidecrypto.com/api/v2/queries/544503b2-6577-49e8-bcf0-d25ebefc5d51/data/latest")
sdf = pd.DataFrame(settled.json())
print(sdf.tail())
```

```python
df_filter = df[df["MAX_PRICE"] < 2500]
df_filter = df_filter[df_filter["DAY_DATE"] > '2023-06-01']
df_filter = df_filter[df_filter["MIN_PRICE"] > 1000]
fig = go.Figure()
fig.add_trace(go.Candlestick(
    x=df_filter["DAY_DATE"],
    open=df_filter["AVG_PRICE"],
    close=df_filter["AVG_PRICE"],
    high=df_filter["MAX_PRICE"],
    low=df_filter["MIN_PRICE"]
))
fig.show()
```

```python
df_new = df_filter
df_new["UP_STD"] = df_new["AVG_PRICE"].std()/2 + df_new["AVG_PRICE"]
df_new["DOWN_STD"] = 0 - df_new["AVG_PRICE"].std()/2 + df_new["AVG_PRICE"]

fig = go.Figure()
fig.add_trace(go.Candlestick(
    x=df_new["DAY_DATE"],
    open=df_new["AVG_PRICE"],
    close=df_new["AVG_PRICE"],
    high=df_new["UP_STD"],
    low=df_new["DOWN_STD"]
))

sdf = sdf[sdf["DAY_DATE"] > '2023-06-01']

fig.add_trace(go.Scatter(x=sdf["DAY_DATE"],y=sdf["PRICE"]))
fig.show()
```
